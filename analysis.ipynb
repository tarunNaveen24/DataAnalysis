```python
df5 = pd.read_csv('State of Washington - Electric Vehicle Population Data.csv')
df5
```

```python
type(df5.loc[0, 'Vehicle Location'])
type(df5.)
#df5.loc[0, 'Electric Range']
```

```python
df5[['x', 'y']] = df5['Vehicle Location'].str.rsplit(' ', 1, expand=True)
df5['y'] = df5['y'].str.replace(')', '')
df5['y']
df5['x']
df5['x'] = df5['x'].str.replace('POINT ', '').str.replace('(', '')
df5['x']
```

```python
# Convert column to float  
df5['x'] = df5['x'].astype(float)
df5['y'] = df5['y'].astype(float)

# Filter between 2 and 4
df5 = df5[(df5['x'] > -130) & (df5['x'] < -110) & (df5['y'] > 45)]
df5
```

```python
from shapely.geometry import Point
def create_point(row):
    return Point(row['x'], row['y'])

# Create 'Coordinate' column using apply() and lambda function
df5['Coordinate'] = df5.apply(lambda row: create_point(row), axis=1)
df5.loc[0, 'Coordinate']
```

```python
import geopandas as gpd
gdf = gpd.GeoDataFrame(df5, geometry='Coordinate')
gdf
```

```python
gdf_us = gpd.read_file('gz_2010_us_040_00_5m.json')
gdf_us
```

```python
gdf_us['NAME'].unique()
gdf_wa = gdf_us[gdf_us['NAME'] == 'Washington']
gdf_wa
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1)
gdf_wa.plot(ax=ax)
gdf.plot(color='black', ax=ax)
plt.title('State of Washington - Electric Vehicle Population Scatterplot')
```
